```python
#import dxl ax12a module
from dxl_ax12a import AX12a

#setup port and initialize a controller instance
DEVICENAME = 'com3'  # Default COM Port
motor_controller = AX12a(DEVICENAME)

#action modes
modes_pattern = ["write_pos", "write_pos_group","wheel_mode", "write_speed", "read_position", "read_speed","read_load"]
```

```python
# SETUP INITIAL STATE

#connected motor ids
connected_motors = [0,1,2]

#speed of each motor
moving_speeds=[100,120,300]

#initial position of each motor
initial_positions=[500,300,500]

#pass the initial state to the motors
motor_controller.set_initial_state(connected_motors, moving_speeds, initial_positions)
```

```python
#execute function according to the given mode
def execute(mode, _id=None,value=None):
    if _id is not None:
        motor_controller.set_id(_id)

    if mode == modes_pattern[0]:
        motor_controller.enable_torque()
        #Input a target position (0-1023)
        motor_controller.set_position(int(value))
    elif mode == modes_pattern[1]:
        motor_controller.enable_torque()
        #Input a target position (0-1023)
        position=int(value)
        positions=[position]*len(connected_motors)
        motor_controller.set_position_group(positions)
    elif mode == modes_pattern[2]:
        motor_controller.enable_torque()
        #Input [0] False, [1] True : 
        motor_controller.set_wheel_mode(int(value))
        
    elif mode == modes_pattern[3]:
        motor_controller.enable_torque()
        #Input a target speed (CCW > 0-1024,CW > 1024-2047)
        motor_controller.set_moving_speed(int(value))

    elif mode == modes_pattern[4]:
        motor_controller.get_position()
    
    elif mode == modes_pattern[5]:
        motor_controller.get_speed()

    elif mode == modes_pattern[6]:
        motor_controller.get_load()

       
```

```python
# 0:"write_pos", id:[0,1,2], value(0-1023)
# 1:"write_pos_group",set position for all motors, value(0-1023)
# 2:"wheel_mode",id:[0,1,2],value(0:False, 1:True)
# 3:"write_speed",id:[0,1,2],value(joint mode:0-1023, wheel mode:CCW 0-1024,CW 1024-2047)
# 4:"read_position",id:[0,1,2]
# 5:"read_speed",id:[0,1,2]
# 6:"read_load",id:[0,1,2]

import time

action_index=0 #input the action index according to the list above 
action_mode=modes_pattern[action_index]
motor_id:int=0
action_value:int=800

execute(mode=action_mode,_id=motor_id,value=action_value)

moves=0

while(True):
    # move back and forth
    if action_value==800:
        action_value=400
    elif action_value==400:
        action_value=800
    execute(mode=action_mode,_id=motor_id,value=action_value)
    moves+=1
    if moves > 15:
        break

    time.sleep(3)
    
    



```

```python
motor_controller.disable_torque_group()
AX12a.close_port()
```
